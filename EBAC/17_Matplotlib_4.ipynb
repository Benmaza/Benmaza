```python
import pandas as pd
df=pd.read_csv('co2-mm-mlo.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
fechas = df[df['Date']>='2015-01-01']

fig, ax = plt.subplots()
ax.plot(fechas['Date'], fechas['Interpolated'])
ax.axes.xaxis.set_ticks([])
plt.xlabel('Datos mensuales 2015-2020')
plt.ylabel('CO2')
plt.show()
```
