```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data_house = pd.read_csv("kc_house_data.csv")
data_house
```

```python
data_house['Intercept'] = 1
data_house= data_house[['Intercept', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']]
data_house
```

```python
# separar datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split

Xdata = data_house[['Intercept', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']].values
Ydata = data_house['price'].values

X_train, X_test, Y_train, Y_test = train_test_split(Xdata, Ydata, test_size=0.3, random_state=1)
```

```python
X = X_train
Y = Y_train
```

```python
# calcular las transpuestas, la inversa de la matriz y la Beta
XT_X = np.matmul(np.matrix.transpose(X), X)
XT_Y = np.matmul(np.matrix.transpose(X), Y)
XT_X_inv = np.linalg.inv(XT_X)
betas = np.matmul(XT_X_inv, XT_Y)

print("Transpuesta de X multiplicada por X:")
print (XT_X)
```

```python
print("Transpuesta de X multiplicada por Y:")
print (XT_Y)
```

```python
print("Inversa de la transpuesta de X multiplicada por X:")
print (XT_X_inv)
```

```python
print("Betas:")
print (betas)
```

### calcular la suma total de cuadrados, la suma cuadrada explicada
### y la suma de cuadrados residuales

```python
# calculo de la suma de cuadrados total (TSS)
TSS = np.matmul(np.matrix.transpose(Y), Y) - len(Y) * (Y.mean() **2)
TSS
```

```python
# calcular la suma de cuadrados explicada (ESS)
ESS = np.matmul(np.matrix.transpose(Y - np.matmul(X, betas)), (Y - np.matmul(X, betas)))
ESS
```

```python
# Calculo de la Residual Sum of Squares (RSS)
RSS= TSS - ESS
RSS 
```

```python
# calculo de R2
R2 = 1 - (RSS / TSS)
R2
```

```python
# Calculo de R2 ajustado
R2_AJ = 1 - (RSS / (len(Y) - X.shape[1])) / (TSS / (X.shape[0] - 1))
R2_AJ
```
