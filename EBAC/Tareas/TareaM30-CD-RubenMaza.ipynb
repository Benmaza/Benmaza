```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
```

```python
# Se crea un DataFrame con un vino de prueba
vino_test = pd.DataFrame({'Alcohol':14.0, 'Malic_Acid':2.0, 'Ash':2.5, 'Ash_Alcanity':16.0, 'Magnesium':115, 'Total_Phenols':3.0, 'Flavanoids':2.5, 'Nonflavanoid_Phenols':0.4, 'Proanthocyanins':2.0, 'Color_Intensity':9, 'Hue':1, 'OD280':3.5, 'Proline':800}, index=[0])
vino_test
```

```python
# se carga la base de datos de los vinos y se imprime una muestra
vinos = pd.read_csv('wine-clustering.csv')
vinos.sample(5)
```

```python
# eliminar valores nulos y verificar que no existan

vinos = vinos.dropna()
vinos.info()
```

```python
# Separar las características y la variable objetivo
X = vinos.drop(columns=['Alcohol'])
y = vinos['Alcohol']

# Crear y entrenar el modelo KNN
knn = NearestNeighbors(n_neighbors=5)
knn.fit(X)
```

```python
# Se carga el vino de prueba 
vino_test = vino_test[X.columns] 
vino_test = vino_test.iloc[[0]] 
```

```python
# Encontrar los 5 vecinos más cercanos
distances, indices = knn.kneighbors(vino_test)

# Obtener los valores de alcohol de los vecinos más cercanos
alcohol_values = y.iloc[indices[0]]

# Calcular el promedio de alcohol
promedio_alcohol = alcohol_values.mean()

# Mostrar los datos de los 5 vinos más parecidos
vinos_parecidos = vinos.iloc[indices[0]]
print("Datos de los 5 vinos más parecidos:")
print(vinos_parecidos)

print(f'\nEl valor promedio de alcohol de los 5 vinos más parecidos es: {promedio_alcohol}')
```

```python
from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
my_basket = [['bread', 'butter', 'wine', 'bananas', 'coffee', 'carrots'],
    ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
    ['beer', 'chips', 'asparagus', 'salsa', 'milk', 'apples'],
    ['olive oil', 'bread', 'butter', 'tomatoes', 'steak', 'carrots'],
    ['tomatoes', 'onions', 'chips', 'wine', 'ketchup', 'orange juice'],
    ['bread', 'butter', 'beer', 'chips', 'milk'],
    ['butter', 'tomatoes', 'carrots', 'coffee', 'sugar'],
    ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
    ['bread', 'butter', 'ketchup', 'coffee', 'chicken wings'],
    ['butter', 'beer', 'chips', 'asparagus', 'apples'],
    ['tomatoes', 'onion', 'beer', 'chips', 'milk', 'coffee']]
    
my_basket
```

```python
# Usar TransactionEncoder para transformar los datos 
te = TransactionEncoder() 
te_ary = te.fit(my_basket).transform(my_basket) 
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# se aplicará el algoritmo Apriori para encontrar los ítemsets frecuentes
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
# Eliminar los ítemsets vacíos 
frequent_itemsets = frequent_itemsets[frequent_itemsets.itemsets.map(len) >= 0]
# ordenar los ítemsets por soporte
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)
# Mostrar los ítemsets frecuentes
frequent_itemsets
```

```python
# Calcula las reglas de asociación de elementos frecuentes con un umbral de confianza de 0.7

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7, num_itemsets=len(frequent_itemsets))
# Mostrar los resultados 
print(frequent_itemsets) 
print(rules.head(20))
```
