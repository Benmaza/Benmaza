```python
# Cargar librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(np.array([[7,6.5,9.2,8.6,8],
                            [7.5,9.4,7.3,7,7],
                            [7.6,9.2,8,8,7.5],
                            [5,6.5,6.5,7,9],
                            [6,6,7.8,8.9,7.3],
                            [7.8,9.6,7.7,8,6.5],
                            [6.3,6.4,8.2,9,7.2],
                            [7.9,9.7,7.5,8,6],
                            [6,6,6.5,5.5,8.7],
                            [6.8,7.2,8.7,9,7]]),
                  index=('Lucia','Pedro','Ines','Luis','Andres','Ana','Carlos','Jose','Sonia','Maria'),
                  columns=('Matematicas','Ciencias','Español','Historia','EdFisica'))
df
```

```python
# Creacion de la matriz de datos

from numpy import array
from sklearn.decomposition import TruncatedSVD

A=df
A
```

```python
from numpy import diag
from numpy import zeros 
from scipy.linalg import svd
```

```python
# Singular-value decomposition (SVD) de la matriz A

U, s, VT = svd(A)
Sigma = zeros((A.shape[0], A.shape[1]))
Sigma[:A.shape[1], :A.shape[1]] = diag(s)
print('Matriz U:')
print(U)
```

```python
print('Matriz Sigma:')
print(Sigma)
```

```python
print('Matriz VT:')
print(VT)
```

```python
# comprobación que la matriz es la misma
B = U.dot(Sigma.dot(VT))
print(B)
```

```python
# Reduccion de la dimension de la matriz A a 2 dimensiones

n_elements = 2

UReduced = U[:, :n_elements]
print('Matriz U reducida:')
print(UReduced)
```

```python
# Sigma reducida
SigmaReduced = Sigma[:n_elements, :n_elements]
# VT reducida
VTReduced = VT[:n_elements, :]
# Reconstrucción de la matriz A
Areduced = UReduced.dot(SigmaReduced.dot(VTReduced))
print('Matriz A reducida:')
print(Areduced)
```

```python
T2 = UReduced.dot(SigmaReduced)
print('Matriz T2:')
print(T2)
```

```python
# Generacion de la grafica de los alumnos en el mapa de 2 dimensiones

df2 = pd.DataFrame(T2, columns=('a','b'))
x = df2.iloc[:, 0]
y = df2.iloc[:, 1]
x =x.to_numpy()
y =y.to_numpy()

fig, ax = plt.subplots()
ax.set_title('Mapa de alumnos')
ax.scatter(x, y)

for i, txt in enumerate(df.index):
    ax.annotate(txt, (x[i], y[i]))
plt.show()
```

```python
# Reduccion de la dimension de la matriz A a 3 dimensiones

n_elements = 3

UReduced3 = U[:, :n_elements]
SigmaReduced3 = Sigma[:n_elements, :n_elements]
VTReduced3 = VT[:n_elements, :]

Areduced3 = UReduced3.dot(SigmaReduced3.dot(VTReduced3))
print('Matriz A reducida:')
print(Areduced3)
```

```python
T3 = UReduced3.dot(SigmaReduced3)
print('Matriz T3:')
print(T3)
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline

df3 = pd.DataFrame(T3, columns=('a','b','c'))
df3
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('Mapa de alumnos')
xline= df3['b']
yline= df3['c']
zline= df3['a']
ax.scatter3D(xline, yline, zline, c=zline, cmap='Oranges')
```

```python
# Generacion de la grafica 3D
from mpl_toolkits.mplot3d import Axes3D

ax3d = plt.figure().add_subplot(111, projection='3d')
arrayx = np.array(df3['a'])
arrayy = np.array(df3['b'])
arrayz = np.array(df3['c'])
ax3d.scatter(arrayx, arrayy, arrayz, c='b', marker='o')
for i in range(len(arrayx)):
    ax3d.text(arrayx[i], arrayy[i], arrayz[i], '%s' % (str(df.index[i])), size=10, zorder=1, color='k')

plt.title('Mapa de alumnos en 3D')
plt.show()
```
