```python
# Importar Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar documento titanic.csv
df=pd.read_csv('titanic.csv')
df
```

```python
# Creacion de Histograma de frecuencias para la variable edad (Age)
plt.hist(df.Age, bins=20, color='blue', edgecolor='black')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Edad')

# guarda la grafica en un archivo jpg
plt.savefig('histograma.jpg', format='jpg', dpi=300)

plt.show()
```

```python
# Creacion de El diagrama de caja para la variable edad (Age)
plt.boxplot(df.Age, patch_artist=True)
plt.title('Diagrama de Caja de Edad')
plt.xlabel('Frecuencia')
plt.ylabel('Edad')
plt.grid()
# guarda la grafica en un archivo jpg
plt.savefig('Diagcaja.jpg', format='jpg', dpi=300)

plt.show()
```

```python
# Creacion del grÃ¡fico de barras para la variable Pclass
# Cuenta la frecuencia de cada clase y ordena por Ã­ndice
bars = df.Pclass.value_counts().sort_index().plot(kind='bar', color='orange')

# AÃ±adir las etiquetas a cada barra
for bar in bars.patches:
  bars.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
            int(bar.get_height()), ha='center', va='center')

plt.xticks(rotation=0) # Rotar las etiquetas del eje x
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Clase')
plt.grid()
# guarda la grafica en un archivo jpg
plt.savefig('pclass.jpg', format='jpg', dpi=300)

plt.show()
```
