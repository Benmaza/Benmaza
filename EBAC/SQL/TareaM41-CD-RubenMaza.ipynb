```python
import pandas as pd
import sqlite3
```

```python
# Crear la conexión a la base de datos
conn = sqlite3.connect('RH.db')
curr = conn.cursor()
```

```python
# Crear la tabla cargando los datos del archivo CSV
data = pd.read_csv('recursos_humanos.csv')
data
```

```python
# Crear la tabla Detalle
data.to_sql('Detalle', conn, if_exists='replace', index=False)
```

```python
conn.commit()
```

```python
# Promedio de satisfacción de los empleados

curr.execute('SELECT AVG(satisfaction_level) FROM Detalle')
resultados = curr.fetchall()
print(resultados)
```

```python
# Promedio de satisfación de los empleados activos
curr.execute('SELECT AVG(satisfaction_level) AS "AVG Satisfacción de empleados activos" FROM Detalle WHERE left = 0;')
resultados = curr.fetchall()
print(resultados)
```

```python
# Promedio de satisfación de los empleados inactivos
curr.execute('SELECT AVG(satisfaction_level) AS "AVG Satisfacción de empleados Inactivos " FROM Detalle WHERE left = 1;')
resultados = curr.fetchall()
print(resultados)
```

```python
# Promedio de horas mensuales de los empleados con salario bajo
curr.execute("SELECT AVG(average_montly_hours) AS 'Horas promedio mensuales salario bajo' FROM Detalle WHERE salary = 'low';")
resultados = curr.fetchall()
print(resultados)
```

```python
# Promedio de horas mensuales de los empleados con salario medio
curr.execute("SELECT AVG(average_montly_hours) AS 'Horas promedio mensuales salario medio' FROM Detalle WHERE salary = 'medium';")
resultados = curr.fetchall()
print(resultados)
```

```python
# Desplegar los empleados que han sido promovidos en los últimos 5 años y que han dejado la empresa
curr.execute("SELECT * FROM Detalle WHERE Promotion_last_5years = 1 AND left = 1;")
resultados = curr.fetchall()
for row in resultados:
    print(row)
```

```python
# Desplegar los empleados que su evaluación fue mayor a 0.90
curr.execute("SELECT * FROM Detalle WHERE last_evaluation > 0.90;")
resultados = curr.fetchall()
for row in resultados:
    print(row)
```
