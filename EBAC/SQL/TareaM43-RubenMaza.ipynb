```python
import sqlite3
import pandas as pd
import csv
```

```python
conn = sqlite3.connect('EmpleadosTareaM43.db')
curr = conn.cursor()
```

```python
# crear la tabla Person
Person = pd.read_csv('Person.csv')
Person
```

```python
Person.to_sql('Person', conn, if_exists='replace', index=False)
```

```python
# Crear la tabla Department
Department = pd.read_csv('Department.csv')
Department.to_sql('Department', conn, if_exists='replace', index=False)
conn.commit()
```

```python
# Crear la tabla Employee
Payhistory = pd.read_csv('EmployeePayhistory.csv')
Payhistory.to_sql('EmployeePayhistory', conn, if_exists='replace', index=False)
conn.commit()
```

```python
# Crear la tabla EmployeeDepartmentHistory
EmployeeDepartmentHistory = pd.read_csv('EmployeeDepartmen.csv')
EmployeeDepartmentHistory.to_sql('EmployeeDepartmen', conn, if_exists='replace', index=False)
conn.commit()
```

```python
# Ejemplos de consultas con INNER JOIN
curr.execute('''
             SELECT * 
             FROM Person INNER JOIN EmployeePayhistory ON Person.BusinessEntityID = EmployeePayhistory.BusinessEntityID''')
resultado = curr.fetchall()
for fila in resultado:
    print(fila)
```

```python
# Ejemplos de consultas con LEFT JOIN
curr.execute('''
             SELECT * 
             FROM Person 
             LEFT JOIN EmployeePayhistory ON Person.BusinessEntityID = EmployeePayhistory.BusinessEntityID''')
resultado = curr.fetchall()
for fila in resultado:
    print(fila)
```

```python
# Ejemplos de consultas con CASE WHEN
curr.execute('''
    SELECT 
        Person.BusinessEntityID, 
        Person.FirstName, 
        Person.LastName, 
        CASE 
            WHEN EmployeePayHistory.PayFrequency = 1 THEN 'Eventual'
            WHEN EmployeePayHistory.PayFrequency = 2 THEN 'Base'
            ELSE 'Comision'
        END AS FrecuenciaPago
    FROM Person 
    INNER JOIN EmployeePayHistory 
    ON Person.BusinessEntityID = EmployeePayHistory.BusinessEntityID
''')
resultado = curr.fetchall()
for fila in resultado:
    print(fila)
```

```python
# Ejemplo de consulta con subconsultas y ANTIJOIN en MiddleName

curr.execute('''
    SELECT 
        p.BusinessEntityID, 
        p.FirstName, 
        p.MiddleName, 
        p.LastName
    FROM Person p
    LEFT JOIN (
        SELECT DISTINCT MiddleName 
        FROM Person 
        WHERE MiddleName IS NOT NULL
    ) sub ON p.MiddleName = sub.MiddleName
    WHERE sub.MiddleName IS NULL
''')
resultado = curr.fetchall()
for fila in resultado:
    print(fila)
```

```python
# Consulta con INNER JOIN en Person y EmployeeDepartmen
curr.execute('''
    SELECT 
        Person.BusinessEntityID, 
        Person.FirstName, 
        Person.LastName, 
        Department.Name
    FROM Person 
    INNER JOIN EmployeeDepartmen ON Person.BusinessEntityID = EmployeeDepartmen.BusinessEntityID
    INNER JOIN Department ON EmployeeDepartmen.DepartmentID = Department.DepartmentID
''')
resultado = curr.fetchall()
for fila in resultado:
    print(fila)
```

```python
curr.execute('''
    SELECT 
        BusinessEntityID, 
        FirstName, 
        LastName         
    FROM Person 
    WHERE BusinessEntityID IN (
        SELECT BusinessEntityID 
        FROM EmployeeDepartmen
        WHERE ShiftID = 2
    )
''')
resultado = curr.fetchall()
for fila in resultado:
    print(fila)
```
