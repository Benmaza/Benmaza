```python
import csv
```

```python
import sqlite3
conn = sqlite3.connect('listadoclientes.db')
```

```python
cur = conn.cursor()
```

```python
# Leer el archivo Csv
with open('datos_clientes.csv', newline='', encoding='utf-8') as csvfile:
   data = list(csv.reader(csvfile))
data
    
```

```python
len(data)
```

```python
# creacion de la tabla donde se importara los datos
cur.execute("""
            CREATE TABLE listadoclientes (
                CustomerID REAL PRIMARY KEY,
                CustomerName TEXT,
                ContactName TEXT,
                Address TEXT,
                City TEXT,
                PostalCode TEXT,
                Country TEXT,
                PurchaseUSD REAL
            );
""")
```

```python
# Inserccion de los datos en la tabla creada

for renglon in range (1,len(data)):
    cur.execute("""
                INSERT INTO listadoclientes (CustomerID, CustomerName, ContactName, Address, City, PostalCode, Country, PurchaseUSD)
                VALUES (?,?, ?, ?, ?, ?, ?, ?)
                """, data[renglon])
```

```python
# Aplicacion de los cambios
conn.commit()
```

```python
conn.close()
```
