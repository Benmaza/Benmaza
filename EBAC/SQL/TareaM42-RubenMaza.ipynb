```python
import sqlite3
import pandas as pd
```

```python
# Crear la conexi√≥n a la base de datos
conn = sqlite3.connect('RH.db')
curr = conn.cursor()
```

```python
# Crear la tabla cargando los datos del archivo CSV
data = pd.read_csv('recursos_humanos.csv')
data
```

```python
# Crear la tabla Detalle
data.to_sql('Detalle', conn, if_exists='replace', index=False)
```

```python
# ORDER BY
curr.execute('SELECT sales, salary, satisfaction_level FROM Detalle ORDER BY satisfaction_level DESC')
resultados = curr.fetchall()
for r in resultados:
    print(r)
```

```python
# ORDER BY con m√∫ltiples columnas
curr.execute('SELECT salary, number_project, satisfaction_level FROM Detalle ORDER BY number_project ASC, satisfaction_level DESC')
resultados = curr.fetchall()
for r in resultados:
    print(r)
```

```python
# GROUP BY
curr.execute('SELECT sales, AVG (last_evaluation) AS AVG_last_evaluation FROM Detalle GROUP BY sales')
resultados = curr.fetchall()
for r in resultados:
    print(r)
```

```python
# GROUP BY con m√∫ltiples columnas
curr.execute('SELECT sales, salary, AVG (last_evaluation) AS AVG_last_evaluation FROM Detalle GROUP BY sales, salary')
resultados = curr.fetchall()
for r in resultados:
    print(r)
```

```python
# GRUOP BY con HAVING
curr.execute('SELECT sales, AVG (Work_accident) AS AVG_Accidents FROM Detalle GROUP BY sales HAVING AVG_Accidents > 0.15')
resultados = curr.fetchall()
for r in resultados:
    print(r)
```

```python
curr.execute('SELECT sales,  SUM(Work_accident) AS Total_Accidents FROM Detalle GROUP BY sales HAVING SUM(Work_Accident) > 200')
resultados = curr.fetchall()
for r in resultados:
    print(r)
```
