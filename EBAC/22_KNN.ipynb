```python
# cargar documento cancer de mama

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# cargar el dataset
data = pd.read_csv('Breast-cancer.csv')
```

```python
data
```

```python
# eliminar columnas categoricas innecesarias

data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
data.head()
```

```python
M= data[data['diagnosis'] == 'M']
B= data[data['diagnosis'] == 'B']
```

```python
# Diagrama de dispersion

plt.scatter(M.radius_mean, M.texture_mean, color='red', label='Maligno', alpha=0.3)
plt.scatter(B.radius_mean, B.texture_mean, color='green', label='Benigno', alpha=0.3)
plt.xlabel('Radio Promedio')
plt.ylabel('Textura Promedio')
plt.legend()
plt.show()
```

```python
# Cambiar la variable objetivo a numerica
# 0 = benigno, 1 = maligno 

data.diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis]
data
```

```python
# Separar las variables dependientes e independientes

y = data.diagnosis.values
x_data = data.drop(['diagnosis'], axis=1) # eliminar la columna diagnosis de x_data
```

```python
# Normalizacion de los datos para que esten en un rango de 0 a 1
# se usa la formula (x - min(x)) / (max(x) - min(x))

x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
x
```

```python
# Dividir los datos en entrenamiento y prueba

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

```python
# Modelo KNN

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)
print('Con KNN (K=3) la precision es: ', knn.score(x_test, y_test))
```

```python
# Creando matriz de confusion

y_pred = prediction
y_true = y_test

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
cm
```

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(5,5))

sns.heatmap(cm, annot=True, linewidths=0.5, linecolor='red', fmt='.0f', ax=ax)
plt.xlabel('Y pronosticado')
plt.ylabel('Y reales')
plt.show()
```

```python
# Cálculo de la precisión global

correctos = cm[0,0] + cm[1,1]
Incorrectos = cm[0,1] + cm[1,0]

precision = correctos / (correctos + Incorrectos)
print('La precision global es: ', precision)
```

```python
precision_benigno = cm[0,0] / (cm[0,0] + cm[1,0])
print('La precision para el diagnóstico benigno es: ', precision_benigno)
```

```python
precision_maligno = cm[1,1] / (cm[1,1] + cm[0,1])
print('La precision para el diagnóstico maligno es: ', precision_maligno)
```

```python
# Curva ROC
# fpr = tasa de falsos positivos
# tpr = tasa de verdaderos positivos

from sklearn.metrics import roc_curve
from sklearn.metrics import auc

y_score = knn.predict_proba(x_test)
fpr, tpr, threshold = roc_curve(y_test, y_score[:,1])
roc_auc = auc(fpr, tpr)

plt.title('Reiceiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1], [0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('Tasa de verdaderos positivos')
plt.xlabel('Tasa de falsos positivos')
plt.title('Curva ROC por KNN')
plt.show()
```
