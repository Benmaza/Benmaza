# https://datahub.io/core/co2-ppm#data

```python
import pandas as pd
df=pd.read_csv('co2-mm-mlo.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
fechas= df[df['Date'] >='2015-01-01']
fig, ax = plt.subplots()
ax.plot(fechas['Date'], fechas['Interpolated'])
plt.show()
```

```python
fechas= df[df['Date'] >='2015-01-01']
fig, ax = plt.subplots()
ax.plot(fechas['Date'], fechas['Interpolated'])
ax.axes.xaxis.set_ticklabels([])
plt.xlabel('Datos Mensuales de 2015-2020')
plt.ylabel('Emisiones de CO2 en ppm')
plt.show()
```

```python
def grafica_series_tiempo(ejes, x, y, color, etiqueta_x, etiqueta_y):
    ejes.plot(x, y, color=color)
    ejes.set_xlabel(etiqueta_x)
    ejes.set_xlabel(etiqueta_y, color=color)
    ejes.tick_params(axis='y', colors=color)
```

```python
fig, ax= plt.subplots()
grafica_series_tiempo(ax, fechas['Date'], fechas['Interpolated'], 'blue', 
                      'Datos Mensuales de 2015-2020', 'Emisiones de CO2 en ppm')
ax.axes.xaxis.set_ticklabels([])
plt.show()
```

```python
fig, ax= plt.subplots()
grafica_series_tiempo(ax, fechas['Date'], fechas['Interpolated'], 'blue', 
                      'Datos Mensuales de 2015-2020', 'Emisiones de CO2 en ppm')
ax2=ax.twinx()
grafica_series_tiempo(ax2, fechas['Date'], fechas['Trend'], 'red', '', 'CO2 Trend')
ax.axes.xaxis.set_ticks([])
plt.show()
```
