```python
import pandas as pd
import numpy as np
```

```python
datos = pd.read_excel('Online Retail.xlsx')
datos.head(10)
```

```python
# Eliminamos los valores nulos

datos = datos.dropna()
datos.info()
```

```python
# Delimitacion de valores positivos en la columna Quantity

datos_reales = datos[datos['Quantity'] > 0]
datos_reales.info()
```

```python
# Eliminamos las transacciones que no tienen un numero de factura
# y las que no tienen un codigo de producto
# agrupamos por factura y descripcion de producto

basket = (datos_reales.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket
```

```python
# Convertimos los valores a 0 o 1 (si se ha comprado o no)

def binarizacion(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_binarizado = basket.applymap(binarizacion)
basket_binarizado
```

```python
# Filtramos las transacciones que tengan al menos 2 productos
basket_filtrado = basket_binarizado[(basket_binarizado > 0).sum(axis=1) >= 2]
basket_filtrado
```

```python
# aplicacion de apriori
# para localizar los articulos que se compran juntos con mayor frecuencia

from mlxtend.frequent_patterns import apriori
```

```python
# Analisis Apriori con un soporte minimo del 0.03

articulos_frecuentes = apriori(basket_filtrado, min_support=0.03, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
articulos_frecuentes['length'] = articulos_frecuentes['itemsets'].apply(lambda x: len(x))
articulos_frecuentes
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
num_itemsets = len(articulos_frecuentes)
association_rules(articulos_frecuentes, num_itemsets=num_itemsets, metric='confidence', 
                  min_threshold=0.5).sort_values('confidence', ascending=False).reset_index(drop=True)
```

```python
num_itemsets = len(articulos_frecuentes)
association_rules(articulos_frecuentes, num_itemsets=num_itemsets, metric='lift', 
                  min_threshold=1.01).sort_values('lift', ascending=False).reset_index(drop=True)
```
