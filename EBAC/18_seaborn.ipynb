```python
# cargar documento y librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('StudentsPerformance.csv')
df
```

```python
# histograma por pandas
df['math score'].plot.hist()
```

```python
# historgrama por seaborn
sns.histplot(df['math score'])
```

```python
# barra de dispersion por seaborn
sns.displot(df['math score'], kde=True, bins = 10) # kernel density estimation
```

```python
# histograma alternativo por seaborn
sns.displot(df["math score"])
```

```python
# creacion de dispersiÃ³n de datos sin histograma
sns.displot(df['math score'], kind = 'kde', fill = True)
```

```python
# creacion de diagramas de dispersion con recta de regresion
sns.regplot(x = 'reading score', y = 'writing score', data = df)
```

```python
sns.lmplot(x = 'reading score', y = 'writing score', data = df)
```

```python
# grafico de dispersion con recta de regresion por categoria de genero
sns.lmplot(x = 'reading score', y = 'writing score', data=df, hue="gender")
```

```python
# grafico de dispersion con recta de regresion por categoria de genero separadas en columnas
sns.lmplot(x = 'reading score', y = 'writing score', data=df, col="gender")
```

```python
# creacion de grafico de dispersion con recta de regresion por categoria de raza
sns.lmplot(x = 'reading score', y = 'writing score', data=df, col="race/ethnicity")
```

```python
for style in ['dark', 'whitegrid', 'darkgrid', 'ticks']:
    sns.set_style(style)
    sns.displot(df['reading score'])
    plt.title(style)
    plt.show()
```

```python
# remocion de ejes
sns.set_style('ticks')
sns.displot(df['reading score'], kde = True)
sns.despine(left=True, bottom=True) # remueve ejes izquierdo y abajo (lineas negras)
```

```python
# combinacion de parametros con matplotlib

fig, ax = plt.subplots()
sns.histplot(df['writing score'], ax=ax, kde=True)
ax.set(xlabel='Puntaje de escritura', ylabel='Frecuencia')
```

```python
fig, ax = plt.subplots()
sns.histplot(df['writing score'], ax=ax, kde=True)
ax.set(xlabel='Puntaje de escritura', ylabel='Frecuencia', xlim=(0,90), title='Histograma de puntaje de escritura')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)
sns.histplot(df['writing score'], ax=ax0, kde=True) # histograma para TODAS las observaciones
sns.histplot(df.query('gender == "male"')["writing score"], ax = ax1, kde=True) # histograma para los hombres
ax0.set(xlabel='Puntaje de escritura', ylabel='Frecuencia', title='Todos los estudiantes')
ax1.set(xlabel='Puntaje de escritura para hombres', ylabel='Frecuencia', title='Hombres')
ax1.axvline(x=75, label ="Objetivo", linestyle='--', color='red')
ax1.legend()
```

```python
# asignacion de colores al histograma
paleta = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

for p in paleta:
    sns.set_palette(p)
    sns.displot(df['reading score'], kde=True)
    plt.title(p)
    plt.show()
```

```python
fig, ax = plt.subplots()
sns.stripplot(data=df, y='writing score', x='math score', jitter=True, hue='math score', palette='viridis')
ax.axes.xaxis.set_ticks([])
plt.show()
```

```python
fig, ax = plt.subplots()
sns.stripplot(data=df, y='writing score', x='race/ethnicity', jitter=True, hue='race/ethnicity')
plt.show()
```

```python
# diagrama de enjambre
sns.swarmplot(data=df, y='writing score', x='race/ethnicity', hue='race/ethnicity', palette='viridis')
```

```python
# creacion de boxplot (caja y bigote)
sns.boxplot(data=df, y = 'writing score', x= 'race/ethnicity', hue='race/ethnicity', palette='viridis')
```

```python
# diagrama de violin
sns.violinplot(data=df, y='writing score' , x= 'race/ethnicity', hue='race/ethnicity', palette='viridis')
```

```python
# diagrama de lvplot (letter value plot)
sns.catplot(data=df, y='writing score', x= 'race/ethnicity',kind = 'boxen', palette='viridis')
```

```python
# creacion de barplot
sns.barplot(data=df, y='writing score', x='race/ethnicity', palette='viridis', hue='lunch')
```

```python
# creacion de poinplot
sns.pointplot(data=df, y='writing score', x= 'race/ethnicity',hue='lunch', palette='viridis')
```

```python
# creacion de countplot
sns.countplot(data=df, y= 'race/ethnicity',hue='lunch', palette='viridis')
```

```python
# creacion de regresion lineal con regplot con marcador +
sns.regplot(data=df, x='reading score', y='writing score', marker = '+')
```

```python
# creacion de grafico residual 
sns.residplot(data=df, x='reading score', y='writing score')
```

```python
# creacion de grafico de regresion con Regplot y media por valor de x
# muy util para enfatizar tendencias

sns.regplot(data=df, x='reading score', y='writing score', x_estimator=np.mean)
```

```python
# Selecciona solo las columnas numÃ©ricas
df_numerico = df.select_dtypes(include=[np.number])

# Calcula la matriz de correlaciÃ³n
correlacion = df_numerico.corr()

# Crea el mapa de calor
sns.heatmap((correlacion), annot=True)

plt.show()
```

```python
# tabla cruzada
pd.crosstab(df['gender'], df['race/ethnicity'], values=df['math score'], aggfunc='mean')
```

```python
sns.heatmap(pd.crosstab(df['gender'], df['race/ethnicity'], values=df['math score'], aggfunc='mean'), annot=True)
```

```python
# uso de FacetGrid
graf= sns.FacetGrid(df, col= "race/ethnicity", col_order=['group A', 'group B', 'group C', 'group D', 'group E'])
graf.map(sns.boxplot, 'math score')
```

```python
# uso de FactorPlot
# AsegÃºrate de que 'df' es tu DataFrame y que tiene una columna 'math score'
sns.catplot(x='math score', y='race/ethnicity', kind='box', data=df)
```

```python
# FaceGrid con regplot
graf= sns.FacetGrid(df, col= "race/ethnicity")
graf.map(plt.scatter, 'reading score', 'writing score')
```

```python
# graficos tabulares de regresion
sns.lmplot(x='reading score', y='writing score', data=df, col='gender', row='race/ethnicity')
```

```python
# creacion de una grafica conjunta
graf = sns.JointGrid(data=df, x='reading score', y='writing score')
graf.plot(sns.histplot, sns.histplot, kde=True)
```
