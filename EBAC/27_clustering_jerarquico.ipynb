```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
from sklearn.preprocessing import normalize

data_scaled = normalize(data)

# Crear un DataFrame con los datos normalizados con las mismas columnas
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)

#visualizar los datos normalizados
data_scaled.head()
```

```python
# Creacion Dendograma

import scipy.cluster.hierarchy as shc

# Formato de la figura

plt.figure(figsize=(15, 10))
plt.title("Dendograma de los clientes")
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
# Obtener la lista de colores del dendrograma
color_list = dend['color_list']

# Convertir la lista en un conjunto para obtener colores nicos
unique_colors = set(color_list)

# Imprimir los colores nicos
print(unique_colors)
```

```python
# Calcular el nmero ptimo de clusters
num_clusters_optimal = len(unique_colors) - 1
print('Nmero ptimo de clusters:', num_clusters_optimal)
```

```python
# Dibujar el dendrograma con el nmero ptimo de clusters
plt.figure(figsize=(15, 10))
plt.title("Dendograma de los clientes")
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
# visualizar el dataset 
data
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
grupos = cluster.fit_predict(data_scaled)
grupos
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(data_scaled['Milk'], data_scaled['Grocery'], c=cluster.labels_, cmap='viridis')
```

```python
# separacion de los grupos mediante PCA

campos = data_scaled.values
print(campos)
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(campos)
campos = pca.transform(campos)
campos
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(campos[:, 0], campos[:, 1], c=cluster.labels_, cmap='viridis')
```

```python
dataframe = pd.DataFrame(grupos, columns=['grupos'])
dataframe
```

```python
dataframe2 = pd.concat([data, dataframe], axis=1, join='inner')
dataframe2
```

```python
# Mostrar todos los datos sin truncar (limites de filas)
pd.options.display.max_rows = None
dataframe2
```

## Analisis del Caso de Estudiantes

```python
data = pd.DataFrame(np.array([[7, 6.5, 9.2, 8.6, 8],
                              [7.5, 9.4, 7.3, 7, 7],
                              [7.6, 9.2, 8, 8, 7.5],
                              [5, 6.5, 6.5, 7, 9],
                              [6, 6, 7.8, 8.9, 7.3],
                              [7.8, 9.6, 7.7, 8, 6.5],
                              [6.3, 6.4, 8.2, 9, 7.2],
                              [7.9, 9.7, 7.5, 8, 6],
                              [6, 6, 6.5, 5.5, 8.7],
                              [6.8, 7.2, 8.7, 9, 7]]),
                    columns=['Matematicas', 'Ciencias', 'Espaol', 'Historia', 'EdFisica'])
data    
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))
plt.title("Dendograma de los estudiantes")
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
# Obtener la lista de colores optimos del dendrograma 

colores_lista = set(dend['color_list'])
num_clusters_optimal = len(colores_lista) - 1
print('Nmero ptimo de clusters:', num_clusters_optimal)
```

```python
# Dibujar el dendrograma con el nmero ptimo de clusters

plt.figure(figsize=(10, 7))
plt.title("Dendograma de los estudiantes")
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=0.15, color='r', linestyle='--')
```

```python
# Analisis de Grfico con PCA

campos = data_scaled
pca = decomposition.PCA(n_components=2)
pca.fit(campos)
campos = pca.transform(campos)
campos
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
grupos = cluster.fit_predict(data_scaled)
grupos
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(campos[:, 0], campos[:, 1], c=cluster.labels_, cmap='viridis')
```

```python
dataframe = pd.DataFrame(grupos, columns=['grupos'])
dataframe
```

```python
dataframe2 = pd.concat([data, dataframe], axis=1, join='inner')
dataframe2
```
