```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.data'
crime = pd.read_csv(url, header=None, na_values='?')
crime.head()
```

```python
crime[127].describe()
```

```python
plt.figure(figsize=(20, 6))
sns.heatmap(crime.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Remover columnas categoricas
crime.drop([0, 1, 2, 3, 4], axis=1, inplace=True)
crime
```

```python
# Remover filas con valores nulos
crime.dropna(inplace=True)
crime
```

```python
# Verificar si hay valores nulos con el mapa de calor
plt.figure(figsize=(20, 6))
sns.heatmap(crime.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
crime.shape
```

```python
# Dividir el dataset en X y Y
X = crime.drop(127, axis=1)
Y = crime[127]
```

```python
X
```

```python
Y
```

```python
# Division en grupos de entrenamiento (training) y prueba (test)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=1)
```

```python
# construccion del modelo de regresion lineal

from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
```

```python
# Impresion de coeficientes
print("Interception: ", linreg.intercept_)
print("Coeficientes: ", linreg.coef_)
```

```python
# Prediccion

Y_pred = linreg.predict(X_test)
Y_pred
```

### calculo de indicadores de la bondad del ajuste

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
# impresion de indicadores de bondad de ajuste
print("Valor de R2: ", r2_score(Y_test, Y_pred))
print("Error absoluto medio: ", metrics.mean_absolute_error(Y_test, Y_pred))
print("Error cuadratico medio: ", metrics.mean_squared_error(Y_test, Y_pred))
print("Raiz del error cuadratico medio: ", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

## Regresion Ridge

```python
# Alpha = 0
from sklearn.linear_model import Ridge
from sklearn.preprocessing import Normalizer
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Normaliza los datos
scaler = Normalizer()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Ajusta el modelo Ridge
ridgereg = Ridge(alpha=0)
ridgereg.fit(X_train_scaled, Y_train)
Y_pred = ridgereg.predict(X_test_scaled)

# Imprime las mÃ©tricas
print("Valor de R2: ", r2_score(Y_test, Y_pred))
print("Error absoluto medio: ", mean_absolute_error(Y_test, Y_pred))
print("Error cuadratico medio: ", mean_squared_error(Y_test, Y_pred))
print("Raiz del error cuadratico medio: ", np.sqrt(mean_squared_error(Y_test, Y_pred)))
```

```python
# Alpha = 0.1
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import Normalizer
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Normaliza los datos
scaler = Normalizer()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Ajusta el modelo Ridge
ridgereg = Ridge(alpha=0.1)
ridgereg.fit(X_train_scaled, Y_train)
Y_pred = ridgereg.predict(X_test_scaled)

# Imprime las mÃ©tricas
print("Valor de R2: ", r2_score(Y_test, Y_pred))
print("Error absoluto medio: ", mean_absolute_error(Y_test, Y_pred))
print("Error cuadratico medio: ", mean_squared_error(Y_test, Y_pred))
print("Raiz del error cuadratico medio: ", np.sqrt(mean_squared_error(Y_test, Y_pred)))
```

```python
# Examinacion de los coeficientes
print("Interception: ", ridgereg.intercept_)
print("Coeficientes: ", ridgereg.coef_)
```

```python
# Definir rango de prueba para Alpha

alpha_range = 10.**np.arange(-2, 3)
alpha_range
```

```python
# Normaliza los datos
normalizer = Normalizer()
X_train_normalized = normalizer.fit_transform(X_train)
X_test_normalized = normalizer.transform(X_test)


# Ajusta el modelo RidgeCV
ridgeregcv = RidgeCV(alphas=alpha_range, scoring='neg_mean_squared_error')
ridgeregcv.fit(X_train_normalized, Y_train)

# ObtÃ©n el mejor valor de alpha
best_alpha = ridgeregcv.alpha_
print("Mejor valor de alpha: ", best_alpha)
```

```python
# PredicciÃ³n mediante el modelo RidgeCV "Mejor valor de alpha"

Y_pred = ridgeregcv.predict(X_test_normalized)
print("Valor de R2: ", r2_score(Y_test, Y_pred))
print("Error absoluto medio: ", mean_absolute_error(Y_test, Y_pred))
print("Error cuadratico medio: ", mean_squared_error(Y_test, Y_pred))
print("Raiz del error cuadratico medio: ", np.sqrt(mean_squared_error(Y_test, Y_pred)))

```

```python
print("Interception: ", ridgeregcv.intercept_)
print("Coeficientes: ", ridgeregcv.coef_)
```

## RegresiÃ³n Lasso

```python
# Prueba con Alpha = 0.01

from sklearn.linear_model import Lasso
lassoreg = Lasso(alpha=0.01)
lassoreg.fit(X_train, Y_train)

print("Interception: ", lassoreg.intercept_)
print("Coeficientes: ", lassoreg.coef_)
```

```python
# Determinacion del grado de bondad de ajuste

Y_pred = lassoreg.predict(X_test)
print("Valor de R2: ", r2_score(Y_test, Y_pred))
print("Error absoluto medio: ", mean_absolute_error(Y_test, Y_pred))
print("Error cuadratico medio: ", mean_squared_error(Y_test, Y_pred))
print("Raiz del error cuadratico medio: ", np.sqrt(mean_squared_error(Y_test, Y_pred)))
```

```python
from sklearn.linear_model import LassoCV
lassoregcv = LassoCV(n_alphas=100, random_state=1)
lassoregcv.fit(X_train, Y_train)
print("Mejor valor de alpha: ", lassoregcv.alpha_)
```

```python
print("Interception: ", lassoregcv.intercept_)
print("Coeficientes: ", lassoregcv.coef_)
```

```python
Y_pred = lassoregcv.predict(X_test)
print("Valor de R2: ", r2_score(Y_test, Y_pred))
print("Error absoluto medio: ", mean_absolute_error(Y_test, Y_pred))
print("Error cuadratico medio: ", mean_squared_error(Y_test, Y_pred))
print("Raiz del error cuadratico medio: ", np.sqrt(mean_squared_error(Y_test, Y_pred)))
```
